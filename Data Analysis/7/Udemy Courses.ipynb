## 7.Find Out Number of courses per subjects

```python
sns.countplot(
    x=df['subject'],hue=df['subject'],palette='Set2',legend=False,width = 0.6)     # x for x axis ,y-axis between showing vertical and horizantal 
plt.xlabel("Subjects",fontsize= 13)
plt.ylabel("Number Of courses per Subject",fontsize= 13)
plt.xticks(rotation=65)
```

## 14.Display 10 Most Popular Courses As Per Number of Subscribers

```python
df.columns
```

```python
df.sort_values(by = 'num_subscribers',ascending = True).head(10)
```

```python
top_10 = df.sort_values(by = 'num_subscribers',ascending = False).head(10)
```

```python
sns.barplot(x = "num_subscribers", y = "course_title", data = top_10, palette="Set2", legend=False)
plt.xlabel("Number Of Subscribers",fontsize = 20)
plt.ylabel("Course",fontsize = 20)
plt.show()
```

## 15.Find The Course Which Is Having The Highest Number of Reviews.
